```python
# load the data
import pandas as pd
data = pd.read_csv("insurance.csv")
data
```

```python
# check if there is a  row contains null values
data[data.age.isna() | data.sex.isna() | data.bmi.isna() | data.children.isna() | data.smoker.isna() | data.region.isna() | data.charges.isna()]
```

```python
# splitting data into features and label
all_columns = list(data.columns.values)
feature_columns = all_columns[:-1:]
x = data[feature_columns]
y = data[all_columns[-1]]
```

```python
x
```

```python
y
```

```python
# encoding the features to  int vectors
x = pd.get_dummies(x,  columns=["sex" , "smoker" , "region"])
x
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train , y_test = train_test_split(x,y, test_size = 0.3)
print(x_train.shape)
print(x_test.shape)
```

```python
# building the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
model.score(x_test,y_test)
```
